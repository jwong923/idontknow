```python
import pandas as pd

df = pd.read_csv("combined_energy_data.csv", parse_dates=[["Dated","Hours"]], dayfirst=True, index_col="Dated_Hours")
```

```python
df.shape
```

```python
#subset
df1=df.iloc[:,0:34]
df1.head()
```

```python
df1.columns = df1.columns.str.replace(r'\s+', '')
df1.columns = df1.columns.str.replace(r'-','_')
# list(df1.columns.values)
```

```python
# df1['ForecastD_1'] = df1['ForecastD_1'].astype(float)
# df1['ForecastJ'] = df1['ForecastJ'].astype(float)

cols = ['ForecastD_1', 'ForecastJ', 'Gas_TAC']
df1[cols] = df1[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
df1.info()
```

```python
# calculate hourly sum
df1.resample('H').mean()
```

```python
df1.dropna(subset=None, inplace=True)
```

```python
df1.resample('H').mean()
```

```python
df1.to_csv('production_history.csv',index=1)
```
